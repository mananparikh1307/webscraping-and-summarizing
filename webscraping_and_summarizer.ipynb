````python
import threading
import trafilatura
import json
import time

class WebScraper:
    def __init__(self, urls):
        self.urls = urls
        self.lock = threading.Lock()

    def extract_text_from_single_web_page(self, url):
        downloaded_url = trafilatura.fetch_url(url)
        if downloaded_url:
            extracted_content = trafilatura.extract(downloaded_url, with_metadata=True, include_comments=False,
                                                    date_extraction_params={'extensive_search': True, 'original_date': True},
                                                    output_format='json')
            if extracted_content:
                json_output = json.loads(extracted_content)
                return json_output.get('text')

    def save_text_to_file(self, url, text):
        with self.lock:
            with open("scraped.txt", "a", encoding="utf-8") as text_file:
                text_file.write(text + "\n")

    def scrape_url(self, url):
        text = self.extract_text_from_single_web_page(url)
        if text:
            self.save_text_to_file(url, text)

    def scrape_urls_without_multithreading(self):
        start_time = time.time()
        for url in self.urls:
            self.scrape_url(url)
        end_time = time.time()
        time_taken = end_time - start_time
        print(f"Time taken without multithreading: {time_taken} seconds")

    def scrape_urls_with_multithreading(self):
        start_time = time.time()
        threads = []
        for url in self.urls:
            thread = threading.Thread(target=self.scrape_url, args=(url,))
            thread.start()
            threads.append(thread)
        for thread in threads:
            thread.join()
        end_time = time.time()
        time_taken = end_time - start_time
        print(f"Time taken with multithreading: {time_taken} seconds")
        print("All threads completed.")

if __name__ == "__main__":
    urls = [
        "https://en.wikipedia.org/wiki/India",
        # Add more URLs here if needed
    ]

    scraper = WebScraper(urls)
    scraper.scrape_urls_without_multithreading()
    scraper.scrape_urls_with_multithreading()
````

````python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer, pipeline
import torch

from transformers import LlamaTokenizer, LlamaForCausalLM

model_path = "openlm-research/open_llama_7b"

tokenizer = LlamaTokenizer.from_pretrained(model_path,quantization_config=quantization_config)
model = LlamaForCausalLM.from_pretrained(
    model_path,
    torch_dtype=torch.float32,
    device_map='auto',
    quantization_config=quantization_config,
    offload_folder="/content/sample_data/offload_folder",

)
pipeline = pipeline(
    "text-generation", #task
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=4000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
````

````python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0})
````

````python
from langchain import PromptTemplate,  LLMChain

template = """
              Write a concise summary only and nothing else of the following text delimited by triple backquotes.
              Return your response in bullet points which covers the key points of the text.
              ```{text}```
              BULLET POINT SUMMARY:
           """

prompt = PromptTemplate(template=template, input_variables=["text"])

llm_chain = LLMChain(prompt=prompt, llm=llm)

````
